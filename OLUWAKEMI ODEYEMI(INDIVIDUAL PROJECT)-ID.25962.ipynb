```python
years = [2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
```

```python
durations = [103,101,99,100,100,95,95,96,93,90]
```

```python
movie_dict = {'years': years,'durations': durations}
```

```python
print(movie_dict)
```

```python
import pandas as pd
```

```python
durations_df = pd.DataFrame(movie_dict)
```

```python
print(durations_df)
```

```python
import matplotlib.pyplot as plt
year = [2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
durations = [103,101,99,100,100,95,95,96,96,90]
plt.plot(year,durations)
plt.xlabel('years')
plt.ylabel('durations(mins)')
plt.title('Netflix Movie Durations 2011-2022')
plt.show()
```

```python
import pandas as pd
```

```python
netflix_df = pd.read_csv('netflix_data.csv')
```

```python
print(netflix_df.head(5))
```

```python
netflix_df_movies_only = netflix_df[netflix_df['type']== 'Movie']
```

```python
print(netflix_df_movies_only)
```

```python
netflix_movies_col_subset = netflix_df_movies_only[['title','country','genre','release_year','duration']]
```

```python
print(netflix_movies_col_subset.head(5))
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(12,8))
release_year = [2016,2011,2009,2008,2019]
duration = [93,78,80,123,95]
plt.scatter(release_year, duration)
plt.xlabel("year of release")
plt.ylabel("duration")
plt.title("Movie Duration by Year of Release")
plt.show()
```
